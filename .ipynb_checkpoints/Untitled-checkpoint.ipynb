```python
import requests
```

```python
endpoint = 'https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/'
url = '{}data/nrg_ti_oil/A....EU27_2020?startPeriod=YEAR&endPeriod=YEAR&format=SDMX-CSV'.format(endpoint)
```

```python
def fetch_data(url: str):
    response = requests.get(url, timeout=30)
    response.raise_for_status()
    return response.json()
```

```python
for year in range(1990,2025):
    url_parametrized = url.replace('YEAR', str(year))
    print(url_parametrized)
```
