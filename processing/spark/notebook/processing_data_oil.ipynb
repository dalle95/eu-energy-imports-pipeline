```python
schema = types.StructType(
			[
				types.StructField ('DATAFLOW', types.StringType(), True),
				types.StructField('LAST UPDATE', types.StringType(), True),
				types.StructField('freq', types.StringType(), True),
				types.StructField('reporter', types.StringType(), True),
				types.StructField('partner', types.StringType(), True),
				types.StructField('product', types.LongType(), True),
				types.StructField('flow', types.LongType(), True),
				types.StructField('indicators', types.StringType(), True),
				types.StructField('TIME_PERIOD', types.StringType(), True),
				types.StructField('OBS_VALUE', types.DoubleType(), True)
			]
)
```

```python
df_oil = spark.read \
    .option("header", True) \
    .schema(schema) \
    .csv("/data/raw/comext/facts/oil_imports")
```

```python
df_oil.count()
```

```python
df_oil_quantity = df_oil.select('reporter', 'TIME_PERIOD', 'OBS_VALUE').filter(df_oil.indicators == 'QUANTITY_IN_100KG')
```

```python
df_oil_quantity.show(10)
```

```python
df_oil_quantity.write.parquet('/data/processed/comext/facts/oil_imports/quantity', mode = 'overwrite')
```

```python
df_oil_value = df_oil.select('reporter', 'TIME_PERIOD', 'OBS_VALUE').filter(df_oil.indicators == 'VALUE_IN_EUROS')
```

```python
df_oil_value.write.parquet('/data/processed/comext/facts/oil_imports/value', mode = 'overwrite')
```
